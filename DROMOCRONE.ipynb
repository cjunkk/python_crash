```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
distance = np.linspace(0,100,num=100)
```

```python
def diretta(distance,v1):
  return distance/v1
```

```python
def riflessa(distance,v1,h1):
  return np.sqrt(distance**2+4*h1**2)/v1
```

```python
def rifratta(distance,v1,h1,v2):
  return distance/v2+2*h1*(np.sqrt(v2**2-v1**2))/v1/v2
```

```python
v1=500
h1=10
v2=1000
```

```python

plt.plot(distance, diretta(distance, v1), label='Diretta')
plt.plot(distance, riflessa(distance, v1, h1), label='Riflessa')
plt.plot(distance, rifratta(distance, v1, h1, v2), label='Rifratta')

plt.gca().invert_yaxis()

plt.xlabel('Distanza')
plt.ylabel('Tempo')
plt.title('Dromocrone')
plt.legend()
plt.show()
```
